```python
import os
from pychorus import find_and_output_chorus

folder = "/content/drive/Shareddrives/music_folder/audio/band/full"
output_folder = "/content/drive/Shareddrives/music_folder/audio/band/chorus/original"

for filename in os.listdir(folder):
  audio_file = folder + '/' + filename
  output_file = output_folder + '/' + filename[:-4] + "_chorus.mp3"

  chorus_times = find_and_output_chorus(audio_file, output_file)

  print(chorus_times)
```
